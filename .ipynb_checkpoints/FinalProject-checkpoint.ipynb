# Исследование рынка игр

## Описание проекта

Интернет-магазин «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.<br><br>
Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными. Не важно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.<br><br>
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

### Описание данных games.csv

**Name** — название игры

**Platform** — платформа

**Year_of_Release** — год выпуска

**Genre** — жанр игры

**NA_sales** — продажи в Северной Америке (миллионы долларов)

**EU_sales** — продажи в Европе (миллионы долларов)

**JP_sales** — продажи в Японии (миллионы долларов)

**Other_sales** — продажи в других странах (миллионы долларов)

**Critic_Score** — оценка критиков (от 0 до 100)

**User_Score** — оценка пользователей (от 0 до 10)

**Rating** — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

#### Обратим внимание на аббревиатуру tbd в столбцах с рейтингом. Отдельно разберем это значение и опишем, как его обработать

Аббревиатура tbd значит <b>to be determined, to be done.</b> То есть, данные были нарочно не заполнены, так как не определились с рейтингом.

```python
display(games[games['user_score'] == 'tbd'].head())
```

```python
games['user_score'] = pd.to_numeric(games['user_score'],errors='coerce')
```

Когда переобразовал данные в нужный тип используя `convert_dtypes()`  столбец  **user_score** конвертилься в string. Потому что, там был значение tbd. Поэтому здесь мы изпользовали метод `pd.to_numeric()`. И аргумент errors = 'coerce' будеть заменять эти значении на NaN

```python
#Убедимься в результатах
print('-' * 100)
display(games.info())
print('-' * 100)
```

```python
#будеть найти соотношениие пропуски в столбцах
def na_percentage(data):
    columns = data.columns
    temp = {}
    for column in columns:
        percentage = data[column].isna().sum() / len(data)
        temp[column] = percentage
    return temp
def print_na_percentage(temp):
    for key in temp:
        print(key,temp[key],'%')
```

```python
print("Ratio of of na in columns:")
print('-' * 100)
print(print_na_percentage(na_percentage(games)))
print('-' * 100)
```

#### Обработаем пропуски при необходимости:

```python
games['rating'] = games['rating'].fillna('Not Given')
```

```python
games = games.dropna(axis = 0, subset = ['name','year_of_release'])
print('-' * 100)
display(games.info())
print('-' * 100)
```

```python
games['user_score'] = (
    games.groupby(['genre'])['user_score'].transform(lambda x: x.replace(np.nan, value = x.median()) )
)
games['critic_score'] = (
    games.groupby(['genre'])['critic_score'].transform(lambda x: x.replace(np.nan, value = int(x.median())) )
)
```

```python
display(games.info())
```

Здесь мы обработали пропуски. Сначала мы заполняли стольбец `rating` с значением **Not Given**. Если мы будемь сбрасывать пропуски то потеряем 40% данных. Затем сбрасывали пропускные строки в столбцы `name` и `year_of_release`. 0.00012% не будеть строго влият на данных. Затем столбцы `user_score` и `critic_score` заполняли данных медианой группируя каждой жанр. Думаю мы так будемь получать более точные цифры для этого столбцы.

#### Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построем диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. Сформулируем выводы.

```python
for platform in top_platforms:
    temp = games_top_platforms[games_top_platforms['platform'] == platform]
    temp = temp[['total_sales', 'user_score','critic_score']]
    print('-' * 15,platform,'-' * 15)
    pd.plotting.scatter_matrix(temp, figsize = (4,4))
    plt.show()
    print('Корреляция между отзывами критиков и продажами {}: {:.2f}'.format(platform,temp['critic_score'].astype('float64').corr(temp['total_sales'].astype('float64'))))
    print('Корреляция между отзывами пользователей и продажами {}: {:.2f}'.format(platform,temp['user_score'].astype('float64').corr(temp['total_sales'].astype('float64'))))
```

Вывод:
- Практически у всех платформ присутствуем корреляция, между оценками пользователей и критиков к продажам. 
- Оценка пользователей слабо влияет на продажи, чего не скажешь о оценке критиков. Покупатели прислушиваются больше к критикам чем к оценкам других пользователей

## Шаг 4. Составим портрет пользователя каждого региона

```python
#Напишем функции для создания сводных таблиц и отсривоки барплотов
def user_profile(parameter, region, dataset):
    data = (
            dataset
            .pivot_table(index = parameter, values = region,aggfunc = 'sum')
            .sort_values(by = region,ascending = False)
            .head()
    )
    data.plot(kind = 'bar', grid = True)
    plt.title("Топ 5 " + parameter + ' в соответствии ' + region)
    plt.ylabel('Продажи  в млн $')
    plt.show()
```

```python
regions = ['na_sales', 'eu_sales', 'jp_sales']
parameters = ['platform', 'genre','rating']
for parameter in parameters:
    for region in regions:
        user_profile(parameter, region, games)
    print('-' * 50)    
```

## Проведем исследование статистических показателей

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевые гипотезы:

- Нулевая гипотеза H₀: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Альтернативная гипотеза H₁: Средние пользовательские рейтинги платформ Xbox One и PC различаются

```python
x_one = games.query('platform == "XOne"')
pc = games.query('platform == "PC"')
```

```python
# Применим Левене тест чтобы чекать equal_var 
stat, p = st.levene(x_one['user_score'], pc['user_score'])
display(p)
```

```python
#Выполним проверку гипотезы. Будем использовать метод ttest_ind
alpha = .05
#Когда применили левене тест получили очень маленькле значение и поэтому equal_var = False
results = st.ttest_ind(x_one['user_score'], pc['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
action = games.query('genre == "Action"')
sports = games.query('genre == "Sports"')
```

```python
stat1, p1 = st.levene(action['user_score'], sports['user_score'])
display(p1)
```

#### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

Сформулируем нулевые гипотезы:

- Нулевая гипотеза H₀: Средние пользовательские рейтинги жанров Action и Sports одинаковые
- Альтернативная гипотеза H₁: Средние пользовательские рейтинги жанров Action и Sports различаются

```python
alpha = .05
results = st.ttest_ind(action['user_score'], sports['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

- Гипотеза: "Средние пользовательские рейтинги жанров Action и Sports разные". Отвергаем нулевую гипотезу. <br>
